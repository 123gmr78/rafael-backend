## Instrucciones previas a ejecutar el programa:

En este codigo, se desarrolló una calculadora que mejora la anterior, de igual manera, ejecuta las operaciones aritmeticas basicas: SUMA ( + ) RESTA ( - ) MULTIPLICACION ( * ) DIVISION ( / ) POTENCIA ( ** )

En la calculadora anterior se manipularon 2 numeros, en ésta podrás introducir 5 numeros, se validara si la entrada es correcta en caso contrario mandara ERROR para volver a introducir los numeros correctamente y continuar con el proceso aritmético.

Esta Calculadora, al finalizar la operacion elegida, igualmente con el resultado, muestra el tipo de dato que se está introduciendo.

Tendras un MENU para elegir la operación aritmética deseada entre los operadores (+, - , * , /, **) en caso de elegir una opcion incorrecta el programa mandara el ERROR SIGUIENTE: 
"Entrada NO Valida:
 llame mas temprano si tiene dudas...
 El equipo de Backend le agradece usar la calculadora"

```python
"""
Aqui comienza nuestra calculadora:
1.- ingresa los 5 numeros
2.- selecciona del MENU la operación que desees realizar
3.- obén los resultados 

NOTA: Para las operaciones de resta, división y potencia la calculadora solicitará el SUSTRAENDO, DIVISOR y EXPONENTE, 
de ésta forma se ejecutará la instrucción a cada número ingresado.

"""


num1 = float(input("ingresa primer numero entre 1 y 5: "))
while not (num1 >= 1 and num1 <= 5):
    num1 = int(input("ERROR: Ingresa un numero entre 1 y 5: "))
    
num2 = float(input("ingresa segundo numero entre 1 y 5: "))
while not (num2 >= 1 and num2 <= 5):
    num2 = int(input("ERROR: Ingresa un numero entre 1 y 5: "))
    
num3 = float(input("ingresa tercer numero entre 1 y 5: "))
while not (num3 >= 1 and num3 <= 5):
    num3 = int(input("ERROR: Ingresa un numero entre 1 y 5: "))
    
num4 = float(input("ingresa cuarto numero entre 1 y 5: "))
while not (num4 >= 1 and num4 <= 5):
    num4 = int(input("ERROR: Ingresa un numero entre 1 y 5: "))
    
num5 = float(input("ingresa quinto numero entre 1 y 5: "))
while not (num5 >= 1 and num5 <= 5):
    num5 = int(input("ERROR: Ingresa un numero entre 1 y 5: "))
    
print("selecciona del MENU CALCULADORA la operacion que desees realizar: ")

print("1 suma")
print("2 resta")
print("3 multiplicacion")
print("4 division")
print("5 potencia")

menu = input("ingresa la operacion deseada: ")
if menu == "1":
    resultado = num1 + num2 + num3 + num4 + num5
    print("la suma es: ", resultado)
    print(type(resultado))
elif menu == "2":
    A = int(input("Ingresa el numero que quieres restar a los numeros elegidos: "))
    resultado1 = (num1 - A)
    resultado2 = (num2 - A)
    resultado3 = (num3 - A)
    resultado4 = (num4 - A)
    resultado5 = (num5 - A)
    print("la resta es: ", resultado1)
    print("la resta es: ", resultado2)
    print("la resta es: ", resultado3)
    print("la resta es: ", resultado4)
    print("la resta es: ", resultado5)  
    print(type(resultado1))
    print(type(resultado2))
    print(type(resultado3))
    print(type(resultado4))
    print(type(resultado5))
elif menu == "3":
    resultado = num1 * num2 * num3 * num4 * num5
    print("la multiplicacion es: ", resultado)
    print(type(resultado))
elif menu == "4":
    y = int(input("Ingresa el divisor: "))
    resultado1 = (num1 // y)
    resultado2 = (num2 // y)
    resultado3 = (num3 // y)
    resultado4 = (num4 // y)
    resultado5 = (num5 // y)
    
    print("la division es: ", resultado1)
    print("la division es: ", resultado2)
    print("la division es: ", resultado3)
    print("la division es: ", resultado4)
    print("la division es: ", resultado5)
    print(type(resultado1))
    print(type(resultado2))
    print(type(resultado3))
    print(type(resultado4))
    print(type(resultado5))
elif menu == "5":
    x = int(input("Ingresa Exponente: "))
    resultado1 = (num1 ** x)
    resultado2 = (num2 ** x)
    resultado3 = (num3 ** x)
    resultado4 = (num4 ** x)
    resultado5 = (num5 ** x)
    
    print("la potencia es: ", resultado1)
    print("la potencia es: ", resultado2)
    print("la potencia es: ", resultado3)
    print("la potencia es: ", resultado4)
    print("la potencia es: ", resultado5)
    print(type(resultado1))
    print(type(resultado2))
    print(type(resultado3))
    print(type(resultado4))
    print(type(resultado5))
else:
    print("Entrada NO Valida:\n llame mas temprano si tiene dudas... \n El equipo de Backend le agradece usar la calculadora")
```
