## Prototipo de carrito de compras, del codigo origen ahora modificado

```python
"""
ESTE ES MI PROTOTIPO DEL CARRITO DE COMPRAS SOLICITADO EN LA ACTIVIDAD

INSTRUCCIONES: 
A) PRIMERAMENTE, HAY QUE INGRESAR DE ACUERDO AL FORMATO QUE PRESENTA EL CODIGO:
[[CANTIDAD,PRODUCTO], [CANTIDAD, PRODUCTO]]
B) PARA SALIR DEL CARRITO DE COMPRAS SE PULSA LA LETRA "S" NOS MOSTRARA EL RESULTADO DE LA COMPRA Y AGRADECIMIENTO
"""

print("Carrito de compras\n")
print(""" 

B I E N V E N I D O 
Te presentamos el MENU de Productos Disponibles:
1.- Pan 
2.- Jamón 
3.- Crema 
4.- Mayonesa 
5.- Lechuga
6.- Pizza
7.- Refresco
8.- Huevo 

""")

print("Ingresa una lista con los productos en este formato: [[Cantidad,Producto]]")
print("Por ejemplo: [[1,1], [3,2]] significa 1 Pan y 2 Jamón" )


my_shopping_cart = eval(input())

print("\n\nTu carrito es: ")

if type(my_shopping_cart) == list:
    if type(my_shopping_cart[0]) == list:
        for element in my_shopping_cart:
            product = ""
            if element[1] == 1:
                product = "Pan"
            if element[1] == 2:
                product = "Jamón"
            if element[1] == 3:
                product = "Crema"
            if element[1] == 4:
                product = "Mayonesa"
            if element[1] == 5:
                product = "Lechuga"
            if element[1] == 6:
                product = "Pizza"
            if element[1] == 7:
                product = "Refresco"
            if element[1] == 8:
                product = "Huevo"
            print("{}" "{}".format(element[0], product))
    else:
        print("Formato incorrecto, debes ingresar una lista de listas")
else:
    print("Formato incorrecto, debes ingresar una lista de listas")
while True:
    product = input("Ingresa la letra (S para Salir de la compra): ")
    if product == "S":
        print("Gracias por comprar con nosotros..... Regresa Pronto...¡¡¡")
        break
```
